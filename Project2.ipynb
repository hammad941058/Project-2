**Calculating q1,q3 and interquartile range**

```python
q3=df4['temp'].quantile(0.75)
q1=df4['temp'].quantile(0.25)
iqr=q3-q1
iqr
```

**Calculating the zscore of the dataset df4**

```python
zscore=(df4['temp']-df4['temp'].mean())/df4['temp'].std()
zscore
```
